```python
import pandas as pd
# 讀取進出口數據的 CSV 檔案
import_data = pd.DataFrame(pd.read_csv("i9101-212347038.csv",  encoding='big5'))
import_data
```

```python
ticker = "^TWII"
twii_data= yf.Ticker('^TWII')
twii_data=yf.download('^TWII',period='max',interval='1d')
twii_price=twii_data['Close']
twii_price.round(2)
```

```python
combined_data = pd.concat([import_data, twii_price], axis=1).dropna()
# Rename the 'Close' column to 'twii_price' in combined_data
combined_data = combined_data.rename(columns={'Close': 'twii_price'})
correlation = combined_data.corr()
print("相關性矩陣：")
print(correlation)
twii_correlation = correlation["twii_price"]
print("\n與台灣加權指數的相關性：")
print(twii_correlation)
```
